## Series.dt Accessor 
- The .dt accessor in pandas is a convenient way to access datetime-like properties of a pandas Series object.
- The .dt accessor allows you to extract various components of a datetime, such as the year, month, day, hour, minute, second, and many more.

#### Note: It is specifically designed for Series with datetime values (i.e., datetime64 type). 

### Mentioned Below are the Date time related properties of pandas series.dt accessor
- Series.dt.date: Returns the date part of the datetime without the time.
- Series.dt.time: Returns the time part of the datetime without the date.
- Series.dt.timetz: Returns the time part of the datetime with the time zone information.
- Series.dt.year: The year of each datetime.
- Series.dt.month: The month of each datetime (1 = January, 12 = December).
- Series.dt.day: The day of the month for each datetime.
- Series.dt.hour: The hour of each datetime.
- Series.dt.minute: The minute of each datetime.
- Series.dt.second: The second of each datetime.
- Series.dt.microsecond: The microsecond of each datetime.
- Series.dt.nanosecond: The nanosecond of each datetime.
- Series.dt.dayofweek: The day of the week with Monday=0, Sunday=6.
- Series.dt.weekday: Alias for dayofweek.
- Series.dt.day_name(): Returns the name of the day of the week.
- Series.dt.week: The week ordinal of the year.
- Series.dt.weekofyear: Alias for week (deprecated).
- Series.dt.dayofyear: The ordinal day of the year.
- Series.dt.quarter: The quarter of the date.
- Series.dt.is_month_start: Indicates if the date is the first day of the month.
- Series.dt.is_month_end: Indicates if the date is the last day of the month.
- Series.dt.is_quarter_start: Indicates if the date is the first day of the quarter.
- Series.dt.is_quarter_end: Indicates if the date is the last day of the quarter.
- Series.dt.is_year_start: Indicates if the date is the first day of the year.
- Series.dt.is_year_end: Indicates if the date is the last day of the year.
- Series.dt.is_leap_year: Indicates if the date belongs to a leap year.
- Series.dt.days_in_month or Series.dt.daysinmonth: The number of days in the month.
- Series.dt.tz: The time zone of the datetime.
- Series.dt.freq: The frequency object if it is set, otherwise None.

# String to Date Time Conversion

### pandas.to_datetime() syntax:
#### pandas.to_datetime(arg, errors=’raise’, dayfirst=False, yearfirst=False, utc=None, box=True, format=None, exact=True, unit=None, infer_datetime_format=False, origin=’unix’, cache=False)
#### Parameters:
- arg: An integer, string, float, list or dict object to convert in to Date time object.
- dayfirst: Boolean value, places day first if True.
- yearfirst: Boolean value, places year first if True.
- utc: Boolean value, Returns time in UTC if True.
- format: String input to tell position of day, month and year.
- Return type: Datetimeme

#### Note:
The default date format, when none is specified, is typically Month-Day-Year (i.e., mm/dd/yyyy).It handles most of the cases even though it is not mentioned in the correct format

To determine whether a given date is valid:
1) Identify the Year:

  - If there is a four-digit value, it is considered the year.
  - If there isn't a four-digit value, select the value that cannot be valid for both the month and the day (i.e., a value greater than 12   is not a valid month).
2) Identify the Month and Day:
  - Of the remaining two values, one should be less than or equal to 12 (indicating it can be a valid month).
  - The other value should be in the correct range for that month and year (indicating a valid day).

### String conversion to Datetime with dayfirst argument

Dayfirst Argument is used to avoid the ambiguity between DD/MM/YYYY and MM/DD/YYYY and should be used as follows:
* If your date format is DD/MM/YYYY, set dayfirst=True.
* If your date format is MM/DD/YYYY, set dayfirst=False.
  
#### Note:
Since Pandas tries to infer the format of the date based on the data provided, but it generally assumes the MM/DD/YYYY format when there is ambiguity between DD/MM/YYYY and MM/DD/YYYY. This is particularly true if the day value is less than or equal to 12, as it could be interpreted as a month.
- To avoid ambiguity and ensure the correct format is used, you should specify the dayfirst parameter

#### Orignial Data Frame

```python
df=pd.DataFrame({'date1':['2022-05-04','2023-06-05','2024-07-06'],#YYYY-MM-DD
                 'date2':['05/04/2022','06/05/2023','07/06/2024'],#MM/DD/YYYY
                 'date3':['05/04/22','06/05/23','07/06/24'],#DD/MM/YY
                 'date4':['04-May-2022','05-June-2023','06-July-2024'],#Day-Month-Year
                 'date5':['May/04/22','Jun/05/23','Jul/06/24'],#Mon/DD/YY
                 'datetime':['2022-05-04 14:30:00','2023-06-05 15:30:00','2024-07-06 16:30:00'],#YYYY-MM-DD HH:MM:SS
                 })
df
```

#### Dataframe After Conversion

```python
df.apply(pd.to_datetime,dayfirst=True)
```

#### another way of writing it

```python
def conv_to_datetime(x):
    return(pd.to_datetime(x,dayfirst=True))
```

```python
df.apply(conv_to_datetime)
```

### String conversion to Datetime with utc argument

 The utc parameter specifies whether or not to convert the resulting datetime object to Coordinated Universal Time (UTC).

When you use utc=True in pd.to_datetime(), it will:

Convert the date or time to a timezone-aware datetime object in UTC.
- If the input is already timezone-aware, it will convert the input to UTC.
- If the input is naive (i.e., does not have timezone information), it will assume it is in the local timezone and then convert it to UTC.

### String conversion to Datetime with Format argument

- Format argument is used to specify the exact format of the date and time strings that you want to convert into datetime objects.
- This is particularly useful when the input date strings have a consistent format that is not easily recognized or is ambiguous to the automatic parsing algorithm of pandas.

### String conversion to Datetime with Format along with exact argument

- When using format with exact=False, pandas may still parse the dates and ignore extra time information or other deviations, as long as the date portion matches the specified format.
- When using format with exact=True, any deviation from the exact format will result in a ValueError.
  
Example to understand the exact use case of this function is shown below

```python
dates_with_time = ['2024-09-05 12:00:00', '2024-09-06 12:00:00', '2024-09-07 12:00:00']
```

#### without exact=False

```python
try:
    pd.to_datetime(dates_with_time, format='%Y-%m-%d',exact=False)
    print("Dates were parsed successfully.")
    print(pd.to_datetime(dates_with_time, format='%Y-%m-%d',exact=False))
except ValueError as e:
    print(f"Error: {e}")
```

#### without exact=True

```python
# Using format with exact=True
try:
    pd.to_datetime(dates_with_time, format='%Y-%m-%d',exact=True)
    print("Dates were parsed successfully.")
except ValueError as e:
    print(f"Error: {e}")
```

# Date time to String Conversion

strftime is a method in Python used to format datetime objects into readable strings. It allows you to specify the format in which you want to display the date and time.
- format: A string representing the desired format for the datetime object. The format codes are based on Python's strftime directives, such as %Y for the full year, %m for the month, %d for the day, etc.

Refer to the below hyperlink to get information about the format codes for all the Date and Time related properties

https://docs.python.org/3/library/datetime.html#format-codes



## On Individual String

### datetime.strftime() Method

```python
date_string = "2023-12-25 15:30:00"
date_object = datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")
```

```python
print("Input Format : YYY-MM-DD hh:mm:ss")
print("To Format - Day, DD Month YYYY :",date_object.strftime("%A, %d %B %Y"))
print("To Format - DD/MM/YY hh:mm",date_object.strftime("%d/%m/%Y %I:%M %p"))
```

```python
def Format_date(date_string):
    # Convert the string to a datetime object
    date_object = datetime.strptime(date_string, "%Y-%m-%dT%H:%M:%S")
    # Format the datetime object into a new string format
    return date_object.strftime("%A, %d %B %Y at %I:%M %p")
```

```python
df['datetime'].apply(Format_date)
```
