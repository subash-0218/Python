# for-else

```python
for index, i in enumerate(x):
    if (index==2):
        continue
    print(index,i)
 
else:
    print("all the iterations ran successfully ")
```

```python
x= ['a','b','c','d']
index =-1
while (index < 3):
    index = index + 1
    if (index==2):
        continue
    print(index,x[index])

else:
    print("all the iterations ran successfully ")   
```
