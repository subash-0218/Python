### pandas.DataFrame.drop


### DataFrame.drop(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')


1. Drop specified labels from rows or columns.
2. Remove rows or columns by specifying label names and corresponding axis, or by directly specifying index or column names. 
3. When using a multi-index, labels on different levels can be removed by specifying the level. 


```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
df.drop(['B', 'C'], axis=1)
```

```python
df.drop(columns=['B', 'C'])
```

```python
df.drop([0, 1])
```

```python
midx = pd.MultiIndex(levels=[['llama', 'cow', 'falcon'],
                             ['speed', 'weight', 'length']],
                     codes=[[0, 0, 0, 1, 1, 1, 2, 2, 2],
                            [0, 1, 2, 0, 1, 2, 0, 1, 2]])
df = pd.DataFrame(index=midx, columns=['big', 'small'],
                  data=[[45, 30], [200, 100], [1.5, 1], [30, 20],
                        [250, 150], [1.5, 0.8], [320, 250],
                        [1, 0.8], [0.3, 0.2]])
df
```

```python
df.drop(index=('falcon', 'weight'))
```

```python
df.drop(index='cow', columns='small')
```

```python
df.drop(index='length', level=1)
```
