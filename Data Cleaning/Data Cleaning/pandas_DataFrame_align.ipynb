### DataFrame.align(other, join='outer', axis=None, level=None, copy=None, fill_value=None, method=_NoDefault.no_default, limit=_NoDefault.no_default, fill_axis=_NoDefault.no_default, broadcast_axis=_NoDefault.no_default)

1. Align two objects on their axes with the specified join method.
2. 
Join method is specified for each axis Index.

### Basic Align

### default join='outer',axis=0

### Align DataFrame and Series

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data, index=['x', 'y', 'z'])

s = pd.Series([7, 8, 9], index=['y', 'z', 'w'])
aligned_df, aligned_s = df.align(s, axis=0)
aligned_df, aligned_s = df.align(s, axis=1)
print("Aligned DataFrame:")
print(aligned_df)
print("\nAligned Series:")
print(aligned_s)
```

### Align DataFrame and DataFrame

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                   'B': ['B0', 'B1', 'B2']},
                  index=[0, 1, 2])

df2 = pd.DataFrame({'A': ['A2', 'A3', 'A4'],
                   'B': ['B2', 'B3', 'B4']},
                  index=[2, 3, 4])


df1_aligned, df2_aligned = df1.align(df2)#default axis=0

print(df1_aligned)
print(df2_aligned)
```

### Aligning with Different Axes

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6'],
                   'D': ['D4', 'D5', 'D6'],
                   'E': ['E4', 'E5', 'E6']})

df1_aligned, df2_aligned = df1.align(df2, axis=1)

print(df1_aligned)
print(df2_aligned)
```

### Specifying Join Type

```python
import pandas as pd


df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                   'B': ['B0', 'B1', 'B2']},
                  index=[0, 1, 2])

df2 = pd.DataFrame({'A': ['A2', 'A3', 'A4'],
                   'B': ['B2', 'B3', 'B4']},
                  index=[2, 3, 4])

df1_aligned, df2_aligned = df1.align(df2, join='inner')

print(df1_aligned)
print(df2_aligned)
```

### Aligning with Different Axes

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A3', 'A5'],
                   'B': ['B0', 'B1', 'B3', 'B5'],
                   'C': ['C0', 'C1', 'C3', 'C5']},
                  index=[0, 1, 3, 5])


df2 = pd.DataFrame({'A': ['A2', 'A4', 'A6'],
                   'B': ['B2', 'B4', 'B6'],
                   'D': ['D2', 'D4', 'D6']},
                  index=[2, 4, 6])


df1_aligned, df2_aligned = df1.align(df2, join='inner', axis=1)

print(df1_aligned)
print(df2_aligned)
```

```python
a1, a2 = df1.align(df2, join='left', axis=1)
print(a1)
print(a2)
```

```python
a1, a2 = df1.align(df2, join='right', axis=1)
print(a1)
print(a2)
```

### Filling Missing Values on Alignment

```python
import pandas as pd


df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                   'B': ['B0', 'NaN', 'B2']},
                  index=[0, 1, 2])

df2 = pd.DataFrame({'A': ['A2', 'A3', 'A4'],
                   'B': ['B2', 'B3', 'B4']},
                  index=[2, 3, 4])

df1_aligned, df2_aligned = df1.align(df2, fill_value='missing')

print(df1_aligned)
print(df2_aligned)
```
