# rename()

## To rename the columns or index labels of a DataFrame. 

### DataFrame.rename(mapper=None, *, index=None, columns=None, axis=None, copy=None, inplace=False, level=None, errors='ignore')

```python
import pandas as pd
data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}
df = pd.DataFrame(data)
```

### Renaming columns using a dictionary

```python
df.rename(columns={'A': 'Col_A', 'B': 'Col_B'}, inplace=True)   # this is mapping
print(df)
```

```python
# data type
print(type(df))
```

### Renaming rows

```python
print(df.rename(index={0: "x", 1: "y"}))
```

### Using axis

```python
print(df.rename(str.lower, axis=1)) # column rename
```

```python
print(df.rename({1:2,2:4},axis=0))  # index rename
```
