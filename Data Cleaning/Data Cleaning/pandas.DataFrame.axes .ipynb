# axes

### Return a list representing the axes of the DataFrame.

```python
import pandas as pd
data = {'col1': [1, 2, 3], 'col2': ['a', 'b', 'c']}
df = pd.DataFrame(data, index=['x', 'y', 'z'])
df
```

```python
axes_list = df.axes
print("Row Labels:", axes_list[0])
print("Column Labels:", axes_list[1])
```
