## eval()

## evaluates a string expression on the DataFrame columns.It allows you to perform calculations, create new columns, and filter data using string expressions.Operates on columns only, not specific rows or elements. 

```python
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df
```

```python
result = df.eval('C = A + B')    # returns a new dataframe
print(result)
```

```python
df.eval('C=A+B',inplace=True)   # using inplace we make changes in existing dataframe
df
```

### Multiple columns can be assigned

```python
df.eval('''
        D=C+A
        E=D-A
        ''',inplace=True)
```
