### Syntax:DataFrame.applymap(func, na_action=None, **kwargs)


```python
1. Apply a function to a Dataframe elementwise.
2. This method applies a function that accepts and returns a scalar to every element of a DataFrame.
```

### Parameters:
func:returns a single value from a single value.

na_action:{None, ‘ignore’}, default None
If ‘ignore’, propagate NaN values, without passing them to func.

**kwargs:Additional keyword arguments to pass as keywords arguments to func.

### Returns:DataFrame

```python
import pandas as pd

def make_big(x):
  return x.upper()

data = {
  "name": ["Sally","Mary","John"],
  "city": ["London", "Tokyo", "Madrid"]
}

df = pd.DataFrame(data)

newdf = df.applymap(make_big)

print(newdf)
 
```

```python
apply() is used to apply a function along a specific axis of the DataFrame (rows or columns).
It works on both Series (columns) and DataFrames.
When applied to a DataFrame, you can specify the axis along which the function should be applied: axis=0 for applying the function to each column, and axis=1 for applying it to each row.
It is typically used to perform complex operations that require access to the entire row or column.
The function passed to apply() operates on 1D arrays (Series) or 2D arrays (DataFrame).
```

```python
applymap():

applymap() is a DataFrame method specifically designed to operate element-wise on each element of the DataFrame.
It works only on DataFrames and not on Series objects.
It applies a function to each element of the DataFrame independently.
It is useful when you want to apply a function to every element in the DataFrame without considering rows or columns.
It's generally faster than using apply() when the operation is element-wise because it avoids the overhead of axis alignment.
```
