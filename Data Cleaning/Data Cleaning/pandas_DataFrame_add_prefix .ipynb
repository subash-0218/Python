### DataFrame.add_prefix(prefix, axis=None)

1. Prefix labels with string prefix.
2. For Series, the row labels are prefixed.
3. For DataFrame, the column labels are prefixed.

### Parameters:
prefix:str,The string to add before each label.

axis{0 or ‘index’, 1 or ‘columns’, None}, default None
Axis to add prefix on

### Returns:
New Series or DataFrame with updated labels.

```python
import pandas as pd
s = pd.Series([1, 2, 3, 4])
s
```

```python
s.add_prefix('item_',axis=0)
```

```python
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [3, 4, 5, 6]})
df
```

```python
df.add_prefix('col_',axis=0)
df.add_prefix('col_',axis=1)
```
