# drop_duplicates

#### The drop_duplicates() function in Pandas is used to remove duplicate rows from a DataFrame. It helps in identifying and eliminating rows where all the values across columns are identical or where specified columns have duplicate values.

#### Arguments
#### subset=None, keep='first', inplace=False, ignore_index=False

```python
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 2, 2, 3, 4],
        'B': ['x', 'y', 'y', 'z', 'z'],
        'C': ['a', 'b', 'b', 'c', 'c']})
```

```python
df
```

#### Using Argument: subset
subset: Columns to consider for identifying duplicates.None is given as default which means it considers all columns.Returns a dataframe


```python
# Removing duplicates based on a subset of columns (A and B)
df.drop_duplicates(subset=['A', 'B'])
```

#### Using Argument: keep
keep: Specifies which duplicate to keep. Returns a dataframe

#### Options are, 
1)'first' (default)
2)'last' 
3) False - drops all duplicates

```python
df1 = pd.DataFrame({'A': [1, 2, 2, 3, 4],
        'B': ['x', 'y', 'y', 'z', 'z']})
```

```python
df1
```

```python
# Removing duplicates and keeping the last occurrence
df1.drop_duplicates(keep='last')
```

```python
# Removing duplicates and keeping the last occurrence
df.drop_duplicates(keep='first')
```

```python
# Removing duplicates and keeping the last occurrence
df.drop_duplicates(keep=False)
```

#### Using argument: inplace
inplace: If True, performs the operation in the original dataframe and it returns 'None'.

```python
df3 = pd.DataFrame( {'A': [1, 2, 2, 3, 3, 3, 4],
        'B': ['x', 'y', 'y', 'z','z','z', 'z']})
```

```python
df3
```

```python
# Removing duplicates in place
df3.drop_duplicates(inplace=True)

# it returns none
```

```python
df3
```

#### Using argument: ignore_index
ignore_index: If True, the resulting DataFrame will have a new RangeIndex,ignoring the original dataframe index.
Returns a dataframe

```python
df4 = pd.DataFrame( {'A': [1, 2, 2, 3, 3, 3, 4],
        'B': ['x', 'y', 'y', 'z','z','z', 'z']})
```

```python
df4
```

```python
# Removing duplicates and resetting index
df4.drop_duplicates(ignore_index=True)
```
