## pandas.DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), by_row='compat', engine='python', engine_kwargs=None, **kwargs)

1. apply function along axis of dataframe(entire row/col,single row/col)
2. obj passed to func=series objects(dataframe_axis=0(columns)/dataframe_axis=1(rows))
3. return type:series/dataframe

## Applying a Function to DataFrame Elements

```python
#built_in_fns
import pandas as pd
import numpy as np
df = pd.DataFrame([[4, 9],[2,5]], columns=['A', 'B'])
#df
df.apply(np.square)
```

```python
#user_defined_fns
import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})
def square(x):
    return x * x
df.apply(square)
```

## apply() with lambda

```python
df.apply(lambda x: x * x)
```

## apply() along axis

```python
# (axis=0):across_all_cols(default)
# (axis=1):across_all_rows

import pandas as pd
import numpy as np
df = pd.DataFrame([[4, 9],[2,5]], columns=['A', 'B'])
#df
df.apply(np.square,axis=0)#across_all_col
#df.apply(np.square,axis=1)#across_all_row
```

## apply() with arguments

```python
import pandas as pd


def sum(x,y,z):
    return x+y+z
df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})
#df
df.apply(sum,args=(1, 2))
```

```python
import pandas as pd


def sum(x,y,z,**kwargs):
    return x+y+z+kwargs
df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})
#df
df.apply(sum,args=(1, 2),c=1,d=1)
```

```python
import pandas as pd

def sum_with_kwargs(x, y, z,t, **kwargs):
    return x + y + z +t+ kwargs['c'] + kwargs['d']
    #args cant have same var name as kwargs dict key

df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})

result = df.apply(sum_with_kwargs, args=(1, 2, 1), c=1, d=1)
print(result)
```

```python
import pandas as pd


def sum(x,y,z):
    return x+y+z
df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})
#df
df.apply(sum,**kwargs={'c':1,'d':2})
```

## apply() with positional and keyword arguments

```python
#positional arg:args=(1,2)
#keyword arg:m=10
import pandas as pd
def sum(x, y, z, m):
    return (x + y + z) * m


df = pd.DataFrame({'A': [1, 2], 'B': [10, 20]})
df.apply(sum, args=(1, 2), m=10)
```

```python
import pandas as pd

# Define a custom function to be applied
def custom_function(x, a=1, b=1):
    return (x + a) * b

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [10, 20, 30]})

# Apply the custom function to each column of the DataFrame
result = df.apply(custom_function, axis=0, raw=False, result_type='expand', args=(2,), by_row='compat', engine='python', engine_kwargs=None, a=10, b=5)

print(result)
```

```python
import pandas as pd

# Create the DataFrame
s1 = pd.DataFrame([[4, 9]], columns=['A', 'B'])

# Define the custom function to sum elements along with additional arguments
def sum_with_args(x, *args):
    return sum(x) + sum(args)

# Apply the custom function to each element of the DataFrame
result = s1.apply(lambda x: sum_with_args(x, 1, 2, 3, 4, 5))

print(result)
```

## Set to true if the row/column should be passed as an ndarray object

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Define a custom function to be applied
def custom_function(row):
    return row.sum()

# Apply the custom function with raw=False (default)
result_default = df.apply(custom_function)
print("With raw=False:")
print(result_default)

# Apply the custom function with raw=True
result_raw = df.apply(custom_function, raw=True)
print("\nWith raw=True:")
print(result_raw)
```

## Specifies how the result will be returned

```python
import pandas as pd

df = pd.DataFrame({'A': [1,9, 3], 'B': [4, 5, 6]})

def custom_function(x):
    return  x*2

result = df.apply(custom_function, axis=1, result_type='expand')
print(result)
```

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 8, 3], 'B': [4, 5, 6]})

def custom_function(x):
    return [x*2]

result = df.apply(custom_function, axis=1, result_type='reduce')
print(result)
```
