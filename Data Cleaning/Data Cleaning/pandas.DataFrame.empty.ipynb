# empty

## It is a boolean attribute that indicates whether a DataFrame is entirely empty or not.

### DataFrame with data

```python
import pandas as pd
import numpy as np

data = {'col1': [1, 2, 3], 'col2': ['a', 'b', 'c']}
df = pd.DataFrame(data)
print(df.empty)    # No Nan values
```

### DataFrame with nan

```python
data = {'col1': [1, 2,np.nan], 'col2': ['a', 'b', 'c']}
df = pd.DataFrame(data)
print(df.empty)   # contains NaNs, not considered empty
```

### DataFrame with only nans

```python
data = {'col1': [np.nan, np.nan], 'col2': [np.nan, np.nan]}
df = pd.DataFrame(data)
print(df.empty)  # contains NaNs, not considered empty
```

### Empty DataFrame

```python
empty_df = pd.DataFrame()
print(empty_df.empty)   # here the dataframe doesn't contain any items
```
