## DataFrame.all(axis=0, bool_only=False, skipna=True, **kwargs)

## The all() method returns one value for each column, True if ALL values in that column are True, otherwise False

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'c1': [True, True,np.nan], 'c2': [False, False,1]})
df
```

```python
df.all(bool_only=False,skipna=False)
```

```python
df.all(axis=1,bool_only=True)
```

```python
df.all(axis=0)
```

```python
import pandas as pd
df_1=pd.DataFrame([True,True])
df_2=pd.DataFrame([False,False])
df_3=pd.DataFrame([True,False])
df_4=pd.DataFrame([True,False],[True,False])
print(df_1.all())
print(df_2.all())
print(df_3.all())
print(df_4.all())
```

## null values

```python
df= pd.DataFrame({
    'A': [True, False, np.nan],
    'B': [False, np.nan, True],
    'C': [np.nan, np.nan, np.nan]
})
df
#df.all(skipna=False)

```

## all() method by default treats NaN (Not a Number) as True (NaN is not equal to zero)

```python
df.all(skipna=True)
```

```python
df_with_nan = pd.DataFrame({
    'A': [True, False, np.nan],
    'B': [False, np.nan, True],
    'C': [np.nan, np.nan, np.nan]
})

df_all_nan = df_with_nan.all(skipna=False)

print(df_all_nan)
```

## condition

```python
df_numeric = pd.DataFrame({
    'X': [1, 2, 3],
    'Y': [4, 5, 6],
    'Z': [7, 8, 9]
})

df_numeric
```

```python
df_condition = (df_numeric['X'] > 0).all()

print(df_condition)
```

## filter values

```python
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, -3],
        'B': [4, 0, 6],
        'C': [7, 8, 9]}

df = pd.DataFrame(data)
df
```

```python
filtered_df = df[(df > 0).all(axis=1)]

print(filtered_df)
```
