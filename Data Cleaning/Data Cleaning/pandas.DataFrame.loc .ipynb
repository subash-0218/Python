# loc()

## Way to access and modify data based on labels (index and columns).

### Error : Raises KeyError if labels are not found in the index or columns.

```python
import pandas as pd
```

```python
data = {'A': [1, 2, 3, 4],
        'B': [5, 6, 7, 8],
        'C': [9, 10, 11, 12]}
df = pd.DataFrame(data)
df
```

## Getting Values

### Accessing a single element

```python
print(df.loc[0, 'A'])
```

```python
# data type of single element
print(type(df.loc[0, 'A']))
```

### Accessing a single row 

```python
print(df.loc[2])   
```

```python
# data type of single row
print(type(df.loc[2] ))
```

### Accessing a single column

```python
print(df.loc[:,'B'])
```

```python
# data type of single column
print(type(df.loc[:,'B']))
```

### Accessing multiple rows and columns

```python
print(df.loc[0:2, 'A':'B'])
```

```python
### data type of multiple rows and columns
print(type(df.loc[0:2, 'A':'B']))
```

### Accessing alternative rows and columns

```python
print(df.loc[0:4:2])  # row-wise
```

```python
print(df.loc[:,'A':'C':2])  # Column-wise
```

### Boolean list with the same length as the row axis

```python
print(df.loc[[False, True, True, False]])
```

```python
# data type of boolean list
print(type(df.loc[[False, True, True, False]]))
```

### Boolean list with the same length as the column axis

```python
print(df.loc[:,[False, True, False]])
```

```python
# data types of boolean list
print(type(df.loc[:,[False, True, False]]))
```

### Conditional that returns a boolean Series

```python
print(df.loc[df['B'] > 6])
```

```python
# data type of conditional list
print(type(df.loc[df['B']>6]))  
# '  df['B']>6  ' data type is series as it returns in boolean and it is a one-dimensional object holding boolean values.
# '  df.loc[df['B']>6]  ' data type is DataFrame as it is a subset of the original DataFrame.
```

### Condition with column labels specified

```python
print(df.loc[df['B'] > 6, ['A']])
```

```python
# data type of conditional boolean series
print(type(df.loc[df['B'] > 6, ['A']]))
```

### Multiple condition

#### Using &

```python
print(df.loc[(df['A'] > 1) & (df['B'] < 8)])
```

```python
# data type of Multiple condition
print(type(df.loc[(df['A'] > 1) & (df['B'] < 8)]))
```

#### Using |

```python
print(df.loc[(df['A'] > 1) | (df['B'] < 8)])
```

### Callable that returns a boolean Series

```python
df.loc[lambda df: df['C'] == 10]
```

```python
# data type of Callable function
print(type(df.loc[lambda df: df['C'] == 10]))
```

## Setting values

### Set value for all items matching the list of labels

```python
df.loc[[2,3],['B']] = 100
print(df)
```

```python
# data type of set value
print(type(df))
```

### Set value for an entire row

```python
df.loc[0] = 50
print(df)
```

### Set value for an entire column

```python
df.loc[:, 'C'] = 25
print(df)
```

### Set value for rows matching callable condition

```python
df.loc[df['B'] > 35] = 0
print(df)
```

### Add value matching location

```python
df.loc[2, "B"] += 10
print(df)
```
