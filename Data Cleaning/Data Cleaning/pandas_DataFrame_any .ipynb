## DataFrame.any(*, axis=0, bool_only=False, skipna=True, **kwargs)

## Return True whether any element is True

```python
import pandas as pd


df = pd.DataFrame({
    'A': [False, False, True, False],
    'B': [False, False, False, False]
})

df.any(axis=1)
```

```python
df = pd.DataFrame({
    'A': [1, -1, 2, -2],
    'B': [0, 0, 0, 0]
})

(df['A'] > 0).any()
```

```python
conditions = (df['A'] > 0) | (df['B'] > 0) # A > 0 or B > 0
print(df[conditions.any(axis=1)])
```

```python
# Identifying missing values
df['C'] = pd.Series([1, None, 1, None])
missing = df.isnull()
print(missing.any(axis=1))
print(df[missing.any(axis=1)])
```
