# pandas.DataFrame.duplicated


### pandas.DataFrame.duplicated(subset=None, keep='first')


### The duplicated() function in Pandas is used to identify duplicate rows in a DataFrame. It returns a boolean Series indicating whether each row is a duplicate or not.

### Parameters:
subset=None,keep='first'

```python
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 2, 2, 3, 4],
        'B': ['x', 'y', 'y', 'z', 'z'],
        'C': ['a', 'b', 'c', 'c', 'c']})
```

```python
df
```

#### Using Argument: subset
subset: Columns to consider for identifying duplicates.
None is given as default which means it considers all columns.
It returns a boolean Series indicating whether each row is a duplicate or not.

```python
df.duplicated(subset=['B','C'])
```

#### Using Argument: keep
keep: Specifies which duplicate to mark as True. Options are 'first' (default), 'last', or False. 'first' marks the first occurrence as False and subsequent occurrences as True. 'last' marks the last occurrence as False and preceding occurrences as True. If False, all duplicates are marked as True.


```python
df.duplicated(keep='first')
```

```python
df.duplicated(keep='last')
```

```python
df.duplicated(keep=False)
```
