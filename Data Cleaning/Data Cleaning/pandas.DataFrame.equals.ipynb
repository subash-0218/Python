# equals()

## compare two DataFrames and determine if they contain the same elements. NaNs in the same location are considered equal.
## They should have same shape and elements including dtypes

## Returns : True if all elements are the same in both objects, False otherwise.

### Same dtypes

```python
import pandas as pd

# First DataFrame
data1 = {'A': [1, 2, 3], 'B': ['a', 'b', 'c']}
df1 = pd.DataFrame(data1)
print(df1)

# Second DataFrame
data2 = {'A': [1, 2, 3], 'B': ['a', 'b', 'c']}
df2 = pd.DataFrame(data2)
print(df2)
```

### Comparing DataFrames

```python
print(df1.equals(df2))  # Output: True
```

### Different dtypes

```python
# First DataFrame
data1 = {'A': [1.0, 2.0, 3.0], 'B': ['a', 'b', 'c']}
df1 = pd.DataFrame(data1)
print(df1)

# Second DataFrame
data2 = {'A': [1, 2, 3], 'B': ['a', 'b', 'c']}
df2 = pd.DataFrame(data2)
print(df2)
```

```python
print(df1.equals(df2))    # output: False  (different dtypes)
```
