 DataFrame.at_time(time, asof=False, axis=None)

```python
import pandas as pd

# Creating a DataFrame with datetime index
dates = pd.date_range('2024-06-01', periods=5, freq='12h')
df = pd.DataFrame({'values': [10, 20, 30, 40, 50]}, index=dates)
print(df)

# Selecting data at a particular time
result = df.at_time('12:00:00')
print(result)
```

```python
# Selecting data at a particular time along columns
result_columns = df.at_time('12:00:00', axis=0)
print(result_columns)
```

```python
result_columns = df.at_time('12:00:00', axis=1)
print(result_columns)
```
