# Dataframe.Groupby

```python
import pandas as pd
import numpy as np
import warnings
# Suppress all warnings
warnings.filterwarnings("ignore")
```

GroupBy in Pandas is a powerful feature that allows you to group data and perform operations on these groups. This feature is similar to the SQL group by statement. In Pandas, GroupBy can be used to group data in many different ways and perform complex data transformations and analysis.The agg() function is commonly used with groupby() in pandas to apply multiple aggregation functions to grouped data. This allows you to compute summary statistics such as sum, mean, count, etc., for each group.

Common Use Cases
- Aggregating Data: Compute summary statistics (e.g., sum, mean) for grouped data.
- Transforming Data: Perform transformations (e.g., normalization) on grouped data.
- Filtering Data: Filter groups based on a condition.

pandas.DataFrame.groupby syntax 

DataFrame.groupby(by=None, axis=_NoDefault.no_default, level=None, as_index=True, sort=True, group_keys=True, observed=_NoDefault.no_default, dropna=True)upBy object

#### Parameters
- by: mapping, function, str, or iterable
- axis : int, default 0
  
- level : If the axis is a MultiIndex (hierarchical), group by a particular level or levels
  
- as_index : For aggregated output, return object with group labels as the index. Only relevant for DataFrame input. as_index=False is   effectively “SQL-style” grouped output
  
- sort : Sort group keys. Get better performance by turning this off. Note this does not influence the order of observations within each -group. groupby preserves the order of rows within each group.
  
- group_keys : When calling apply, add group keys to index to identify pieces
  
- squeeze : Reduce the dimensionality of the return type if possible, otherwise return a consistent type
  
- Returns : GroupBy object

#### Consider the below example Consisting of The Details of employees 

```python
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 60000, 55000, 70000, 65000, 62000, 58000],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)
df
```

#### Use of agg() function
The agg() function is commonly used with groupby() in pandas to apply multiple aggregation functions to grouped data. This allows you to compute summary statistics such as sum, mean, count, etc., for each group.

```python
df[['Department','Salary']].groupby(by=['Department']).agg(sum)
```

#### use of reset index function

In pandas, when you use groupby(), the grouped columns typically become the index of the resulting DataFrame. The reset_index() function is used to turn the groupby result back into a standard DataFrame by moving the index (the grouped columns) back into regular columns.

```python
df[['Department','Salary']].groupby(by=['Department'],as_index=False).agg(sum)
```

# Understanding the arguments

## "by" Argument

In Python's pandas library, the by argument in the groupby() function is used to specify the key or keys by which the data should be grouped. These keys can be:

- Column names: You can group the data based on one or more columns in the DataFrame.
- A function: You can pass a function that defines how to group the data.
- A list or array: You can pass an external list or array that indicates the group for each row.

```python
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 60000, 55000, 70000, 65000, 62000, 58000],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)
```

#### Original Data Frame 

```python
df
```

#### Single string as input to by argument
You can pass the name of a single column by which the operation should be performed.

```python
df[['Department','Salary']].groupby(by="Department").agg(sum).reset_index()
```

#### List of column names as input to by argument
If you want to perform the operation on multiple columns, you can pass a list of column names.

```python
df[['Department','Gender','Salary']].groupby(by=["Department","Gender"]).agg(sum).reset_index()
```

##  "as_index" argument

- as_index=False: The group labels are not set as the index; instead, they are kept as regular columns in the resulting DataFrame. This is often useful when you want to reset the index and have a more "flat" DataFrame structure, making it easier to work with the grouped results.

# Aggregation functions

Aggregation functions in a groupby operation are used to summarize data by applying a function to groups of data. When you group data by one or more columns, you can use aggregation functions to compute statistics like sums, means, counts, etc., for each group. Some common aggregation functions include:
- sum(): Returns the sum of the values for each group.
- mean(): Returns the mean (average) of the values for each group.
- median(): Returns the median of the values for each group.
- count(): Returns the number of non-null values for each group.
- size(): Returns the size of each group (includes NaN values).
- min(): Returns the minimum value of each group.
- max(): Returns the maximum value of each group.
- std(): Returns the standard deviation of each group.
- var(): Returns the variance of each group.
- prod(): Returns the product of the values for each group.
- describe(): Returns summary statistics for each group (count, mean, std, min, max, etc.).
- agg() or aggregate(): Apply multiple aggregation functions to columns. Can use predefined functions or custom functions.

## DataFrame.Groupby.rank

Methods:
- average: the average rank of group.
- min: lowest rank in group.
- max: highest rank in group.
- first: ranks assigned in order they appear in the array.
- dense: like ‘min’, but rank always increases by 1 between groups.

please refer to the below link to understand more regarding the rank function in groupby:
https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.rank.html

```python
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'HR'],
    'Employee': ['Grace', 'Bob', 'Charlie', 'David', 'Eva', 'Frank','Alice'],
    'Salary': [50000, 60000, 55000, 70000, 65000, 62000,50000],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)
```

#### Original Data Frame 

```python
df
```

```python
df["Average Rank"]=df[['Department','Salary']].groupby(by=["Department"]).rank(method='average')
df["Lowest Rank"]=df[['Department','Salary']].groupby(by=["Department"]).rank(method='min')
df["Highest Rank"]=df[['Department','Salary']].groupby(by=["Department"]).rank(method='max')
df["First Rank"]=df[['Department','Salary']].groupby(by=["Department"]).rank(method='first')
df["Dense Rank"]=df[['Department','Salary']].groupby(by=["Department"]).rank(method='dense')
```

#### The Output for all the different methods available in Rank function
method for ranking refers to how the ranks are assigned when there are tied values(multiple items with the same value) within a group.
1) Average
   - the average method assigns the average of the ranks that would have been assigned to those tied values.
2) min
   - the min method assigns the smallest rank to all tied values.
3) max
   - the max method assigns the highest rank to all tied values.
4) first
   -  the first method assigns ranks based on the order in which the tied values appear in the original data.
5) dense
   - the dense method assigns the same rank to all tied values. The next distinct value gets the next consecutive rank, without skipping any rank values.

```python
df
```

## DataFrame.Groupby.filter

- groupby.filter is used to filter out entire groups based on a condition. It works on data that has been grouped by one or more columns using groupby(), and it evaluates a function for each group. If the group meets the condition, it is kept in the output; otherwise, it is removed.
##### Syntax: 
DataFrame.groupby('column_name').filter(function)

- use the hyperlink for more information
  https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.DataFrameGroupBy.filter.html

#### Application of groupby.filter:
problem:
You manage a chain of retail stores that sell various products across different categories. Each store reports its sales data for each product sold in a particular month. The objective is to identify the best-performing stores for the month, based on the total sales they achieved across all products.The task is to filter out the top-performing stores where the total sales for the month exceed a given threshold.

Solution:
one way to solve this is by using the filter function along with groupby. You can set a sales threshold and then filter out any groups where the total sales exceed that threshold.
consider threshold to be $25000 and the result shows that only store A and store C have the total sales value more than this threshold

```python
sales_data= {
    'Store': ['Store_A', 'Store_A', 'Store_A', 'Store_B', 'Store_B', 'Store_B', 
              'Store_C', 'Store_C', 'Store_C', 'Store_D', 'Store_D', 'Store_D'],
    'Product': ['Product_1', 'Product_2', 'Product_3', 'Product_4', 'Product_5', 'Product_6', 
                'Product_7', 'Product_8', 'Product_9', 'Product_10', 'Product_11', 'Product_12'],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 
                 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing'],
    'Month': ['Sep-2024']*12,
    'Sales': [12000, 8000, 10000, 7000, 6000, 9000, 22000, 15000, 10000, 9000, 8000, 6000]

            }

df = pd.DataFrame(sales_data)
sales_threshold = 25000

# Group by Store and filter based on total sales greater than the threshold
best_stores = df.groupby('Store').filter(lambda x: x['Sales'].sum()>sales_threshold)
best_stores
```

## Cumulative functions

### cumsum

groupby().cumsum() in pandas is used to calculate the cumulative sum of values within each group in a DataFrame. It is an extension of the groupby() functionality, where after grouping data by one or more columns, the cumulative sum (cumsum()) is calculated for each group separately.

#### Practical Use case of Cumsum
##### Problem Statement:
- You are tasked with analyzing the performance of players in the Indian Premier League (IPL) across several matches during the season. You have data showing how many runs each player has scored in each match. Your objective is to determine which players have contributed the most consistently over the season by tracking their performance across matches.

##### solution:
- To solve this, a good approach is to calculate the cumulative total of runs for each player over the course of the season. This allows    you to:
  - Track how many runs each player has scored up to any given match.
  - Compare players based on their overall contributions after each match.
  - Observe trends in performance, such as consistency or improvement over time.

Consider the below example containing the data of player along with the match number along with the number of runs scored correspondingly.

```python
data = {
    'Player': ['Player A', 'Player A', 'Player A', 'Player B', 'Player B', 'Player C'],
    'Match': [1, 2, 3, 1, 2, 1],
    'Runs': [45, 30, 50, 20, 60, 40]
}
df=pd.DataFrame(data)
df
```

to determine the performance accross different matches we can track using cumulative sum,It shows how the player's total runs have accumulated over the season as they play more matches. For example, if a player scores 45 runs in the first match and 30 runs in the second match, the cumulative total after the second match will be 75. This provides insight into the player's consistency and performance throughout the season.

```python
df['Cumulative_Runs'] = df.groupby('Player')['Runs'].cumsum()
df
```

### cumcount

cumcount() is a function in pandas that returns the cumulative count of occurrences of each unique value in a column or a group. It is often used in data manipulation when you need to create a cumulative sequence number for repeated values within a group.
##### Note: The count begins at 0, but you can adjust it to start at 1(by adding 1) if necessary.
#### Practical Use case of Cumcount
##### Problem Statement:
You are working for an e-commerce platform and need to analyze customer purchase behavior. Each customer can place multiple orders, and the dataset contains a record of each order placed by a customer. The goal is to determine how many purchases each customer has made up to and including each transaction to track their engagement and frequency of orders over time.
##### solution:
To solve this, we can use the cumcount() function to generate a running total of purchases for each customer. This will help us track how many times each customer has made a purchase as of any given order.

Assume the below dataset that contains customer IDs and their order dates or order numbers.

```python
data = {
    'Customer_ID': ['C001', 'C001', 'C001', 'C002', 'C002', 'C003'],
    'Order_Date': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-01-18', '2023-03-05', '2023-01-25']
}

df = pd.DataFrame(data)
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df
```

- For this problem statement cumcount() function is highly useful for tracking the purchase behavior of each customer over time. Specifically, it allows you to count how many purchases a customer has made up to and including each transaction.
- By observing how the "Purchase Count" progresses for each transaction, you can see how frequently the customer buys products. If the purchase count increases rapidly, it indicates that the customer is making frequent purchases in a short period.

```python
df['Purchase_Count'] = df.groupby('Customer_ID').cumcount()+1
df
```

### cum min

- This function calculates the cumulative minimum of a column or a Series. It means that at any given index i, the value is the smallest value seen so far in the sequence (from the beginning up to that index).
For example, for a sequence [3, 1, 4, 2], the cumulative minimum would be [3, 1, 1, 1].
- When you use cummin() after groupby(), pandas computes the cumulative minimum within each group independently. So for each group, the cumulative minimum calculation restarts from the first element of the group.
#### Practical Use case of Cummin
##### Problem Statement:
You are a financial analyst responsible for monitoring and reporting on stock prices for your company’s investment portfolio. Your manager has tasked you with developing a method to track the daily stock prices and, more importantly, to identify the lowest price observed up to each day.
##### solution:
We can use the cumulative minimum (cumulative min) function to keep track of the lowest stock price observed up to each date. This means that for each day, we will record the smallest stock price observed so far.

Assume the below dataset that contains stock information of some companies along with Date and price.

```python
data = {
    'Company': ['AAPL', 'AAPL', 'AAPL', 'GOOG', 'GOOG', 'GOOG', 'AAPL', 'GOOG'],
    'Date': ['2023-09-20', '2023-09-21', '2023-09-22', '2023-09-20', '2023-09-21', '2023-09-22', '2023-09-23', '2023-09-23'],
    'Price': [150, 148, 151, 2800, 2750, 2780, 149, 2775]
}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])
df
```

The column cumulative_min_price gives us insights on the variation of price from the minimum price till that point of time to analyze stocks 

```python
df['Cumulative_Min_Price'] = df.groupby('Company')['Price'].cummin()
df
```

### cum max

groupby.cummax is a method in Pandas that computes the cumulative maximum of a column or a set of columns for each group within a groupby object. It returns the maximum value encountered up to the current point in each group.

#### Practical Use case of Cummin
##### Problem Statement:
You are conducting an under water breath holding competition and analyzing the performance of swimmers, specifically tracking how long each swimmer can stay underwater during their training sessions. ## DataFrame.Groupby.apply

In Pandas, groupby.apply is a method used to apply a function to each group of a DataFrame after it has been grouped by one or more columns. This allows for more flexible data manipulation and aggregation compared to standard aggregation methods. 
#### Practical Use case of Groupby.apply
##### Problem Statement:
You are working with an employee dataset that includes details about employees, their departments, salaries, and performance ratings. Your task is to categorize each employee into one of three performance categories based on the following criteria:
- High Performer: An employee is considered a high performer if their performance rating is 4.0 or higher and their salary is above or equal to the median salary of their department.
- Average Performer: An employee is considered an average performer if they do not meet the criteria for high or low performers.
- Low Performer: An employee is considered a low performer if their performance rating is below 4.0 and their salary is below the median salary of their department.
###### Solution:
we can use groupby.apply function to solve these kind of problem statements.First,we create a function to categorize each employee based on their performance rating and salary and then apply that function to each group to perform this group wise. 

```python
import pandas as pd

# Step 1: Create a DataFrame with employee data
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Department': ['HR', 'HR', 'Finance', 'Finance', 'HR', 'Finance'],
    'Salary': [70000, 60000, 90000, 80000, 75000, 85000],
    'Performance Rating': [4.5, 3.0, 4.0, 2.5, 5.0, 3.5]
}

df = pd.DataFrame(data)
df
```

The Function categorize_performance takes each row in a group and assigns the output to the new column performance category based on their performance rating column
- if performance rating is greater than 4 then they are considered as "High Performer" and
- if they have performance rating less than 4 then "Low Performer"
- else an "Average Performer" 

```python
def categorize_performance(group):
    def performance_category(row):
        if row['Performance Rating'] > 4.0: 
            return 'High Performer'
        elif row['Performance Rating'] < 4.0: 
            return 'Low Performer'
        else:
            return 'Average Performer'
    
    group['Performance Category'] = group.apply(performance_category, axis=1)
    return group
```

```python
result = df.groupby('Department').apply(categorize_performance).reset_index(drop=True)
result
```
