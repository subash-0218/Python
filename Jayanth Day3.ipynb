```python
from functools import reduce

#reduce(func, iterable, intial(o))
def fun(x,y):
    return min(x,y)
reduce(fun,[1,2,3,4,5],-1)
```

```python
def fun(x,y):
    return x+y
def average(l):
    return reduce(fun,l)/len(l)
average([1,2,3,4,5])
```
