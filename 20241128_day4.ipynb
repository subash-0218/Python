```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
for i in df['A'].unique().tolist():
    #df.loc[(df['B'].isna()) & (df['A']==i),['B'] ] = df.loc[(df['A']==i),['B']].mean().values[0]
    df.loc[(df['B'].isna()) & (df['A']==i),['B'] ] = df.groupby('A')['B'].mean()[i]
df
```

# apply

```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
def fun(x):
    if (x.dtypes == 'object'):
        return "".join(x.tolist())
    else:
        return x.sum()
df.apply(fun,axis=0)
```

```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
def fun(x):
    #print(x.name,x['A'],x['B'])
    return x['B']+100 
df.apply(fun,axis=1)
df
```

```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
def fun(x):
    x['C'] = x['B']+100
    return x
df.apply(fun,axis=1)
```

```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
def fun(x):
    print(x.mean())
df.groupby('A').apply(fun)
```

```python
df = pd.DataFrame({"A":['a','b','a','a','c','b','c','c','b'],"B":[1,2,np.nan,3,np.nan,5,8,9,np.nan]})
def fun(x,y):
    print(x,y)
df.groupby('A').apply(fun,10)
```
