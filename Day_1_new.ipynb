### Tuples


Tuples in Python are immutable sequences, similar to lists, but they cannot be modified after creation. They are created using parentheses () and can contain elements of any data type, including other tuples.

```python
# Creating a tuple
person = ("John", 25, "London")

# Accessing elements
print(person[0])  # Output: John
print(person[1:])  # Output: (25, 'London')

# Unpacking a tuple
name, age, city = person
print(name)  # Output: John

# Tuples as dictionary keys
point = (3, 4)
coordinates = {point: "Origin"}
print(coordinates)  # Output: {(3, 4): 'Origin'}

#inbuilt functions on tuples
numbers = (1,2,3,2)
print(len(numbers))
print(numbers.count(2))
print(numbers.index(1))

# manipulation of tuples

print(numbers*2)
alphabets =('a','b','c')
combined_tuple = numbers+alphabets
print(combined_tuple)
```

### User-Defined Function

<p>A user-defined function is a reusable block of code that performs a specific task. It allows you to encapsulate functionality, improve code organization, and promote code reusability. In Python, you can define your own functions using the <strong>def</strong> keyword.</p>
<p>The syntax for defining a function is as follows:</p>
<pre><code>def function_name(parameters):
    # Function body
    # Perform desired operations
    # Return a value if needed</code></pre>
<p>Let's break down the components of a user-defined function:</p>
<ul>
    <li><strong>def:</strong> It is a keyword indicating the start of a function definition.</li>
    <li><strong>function_name:</strong> It is the name of the function that you choose. It should follow Python's naming conventions.</li>
    <li><strong>parameters:</strong> These are optional input parameters that the function can accept. You can specify zero or more parameters separated by commas.</li>
    <li><strong>Function body:</strong> It consists of the statements that make up the function's code block. These statements are indented under the function definition.</li>
    <li><strong>Return statement:</strong> It is optional and is used to return a value from the function. If no return statement is provided, the function returns <code>None</code> by default.</li>
</ul>
<p>Here's an example of a user-defined function that calculates the factorial of a number:</p>
<pre><code>def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
num = 5
factorial_num = factorial(num)
print(f"The factorial of {num} is {factorial_num}")</code></pre>
<p>In this example, the function <code>factorial</code> takes an input parameter <code>n</code> and calculates the factorial of that number. The function uses a loop to iterate from 1 to <code>n</code> and multiplies the numbers to compute the factorial. Finally, it returns the result.</p>
<p>User-defined functions can be called multiple times with different arguments, making them highly reusable. They help in modularizing code and promoting code maintainability.</p>
<p>By defining your own functions, you can tailor your code to specific tasks, enhance code readability, and simplify complex operations.</p>

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

# Calling the function with both parameters provided
greet("John", "Hi")

# Calling the function with only one parameter provided
greet("Kate")
```

<h3> Variable Length Arguments </h3>

<p>In Python, *args and **kwargs provide flexibility in accepting a variable number of arguments in function definitions. Understanding when to use *args and **kwargs can help you handle different scenarios effectively.</p>
<h3>Usage of *args</h3>
<ul>
    <li>Use *args when you want to accept a <strong>variable number of positional arguments</strong> (arguments without keywords).</li>
    <li>It allows you to pass any number of arguments to a function, and they will be treated as a <strong>tuple</strong> inside the function body.</li>
    <li>*args is commonly used when:</li>
    <ul>
        <li>You don't know in advance how many arguments will be passed to the function.</li>
        <li>You want to <strong>iterate</strong> over the arguments or pass them to another function that accepts a variable number of arguments.</li>
    </ul>
</ul>
<h3>Usage of **kwargs</h3>
<ul>
    <li>Use **kwargs when you want to accept a <strong>variable number of keyword arguments</strong> (arguments with keywords).</li>
    <li>It allows you to pass any number of keyword arguments to a function, and they will be treated as a <strong>dictionary</strong> inside the function body.</li>
    <li>**kwargs is commonly used when:</li>
    <ul>
        <li>You want to provide <strong>flexibility</strong> for the caller to pass named arguments with values.</li>
        <li>You want to handle <strong>optional</strong> or <strong>additional</strong> arguments that may vary between function calls.</li>
    </ul>
</ul>
<h3>Difference Between *args and **kwargs</h3>
<ul>
    <li>The main difference lies in the type of arguments they handle:</li>
    <ul>
        <li>*args handles <strong>positional arguments</strong> as a tuple.</li>
        <li>**kwargs handles <strong>keyword arguments</strong> as a dictionary.</li>
    </ul>
    <li>With *args, the order of the arguments matters, as they are accessed by their position.</li>
    <li>With **kwargs, the order of the arguments doesn't matter, as they are accessed by their keyword.</li>
    <li>You can use *args and **kwargs together in a function definition to handle both positional and keyword arguments.</li>
</ul>
<p>Understanding when and how to use *args and **kwargs allows you to create more versatile functions that can handle different argument scenarios and enhance the flexibility of your code.</p>

```python
def print_values(*args):
    for arg in args:
        print(arg)

print_values(1, 2, 3)
print_values('apple', 'banana', 'cherry')
```

```python
def print_values(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_values(name='John', age=25)
print_values(country='USA', city='New York')
```

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply(2, 3, 4))  # Output: 24
print(multiply(5, 6))     # Output: 30
```
