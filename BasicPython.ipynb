# Write a class introduction

````python
class MyAttributeClass:
    # Constructor
    def __init__(self, name, age, gender):
        self.name = name  # Create an instance variable
        self.age = age
        self.gender = gender
    
    # Instance method
    def DisplayAttributes(self, bold=False):
        if bold:
            print('HELLO, %s! your age is %d and you happen to be %s' % (self.name.upper(),self.age,self.gender))        
        else:
            print('HELLO, %s! your age is %d and you happen to be %s' % (self.name,self.age,self.gender))
    
    def ReturnDictionary(self):
        return {'name':self.name,'age':self.age, 'gender':self.gender}
    
    def ReturnHTML(self):
        return '<p style="text-align: left;">&nbsp;</p><table style="height: 79px;" width="319"><tbody><tr><td style="width: 151px;"><strong>Name</strong></td><td style="width: 152px;">'+self.name+'</td></tr><tr><td style="width: 151px;"><strong>Age</strong></td><td style="width: 152px;">'+str(self.age)+'</td></tr><tr><td style="width: 151px;"><strong>Gender</strong></td><td style="width: 152px;">'+self.gender+'</td></tr></tbody></table>'
````

````python
Jayanth = MyAttributeClass("gaurav garg",45,'Male') # Construct an instance of class
Jayanth.DisplayAttributes(bold=True) # displays the attributed in capitals
Jayanth.DisplayAttributes()# displays the attributes
print(Jayanth.name)
print(Jayanth.name.upper())
print(Jayanth.age)
print(Jayanth.gender)
print(Jayanth.ReturnDictionary())
print(Jayanth.ReturnHTML())
````
